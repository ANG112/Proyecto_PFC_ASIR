### Funciones

```python
### FUNCIÓN 1: Crea un reporte de la cardinalidad y saca los datos como df ####

def reporte_cardinalidad(df):
    # Calculamos el total de filas una sola vez
    total_filas = len(df)
    
    # Creamos un DataFrame con los resultados
    reporte = pd.DataFrame({
        'Columna': df.columns,
        'Nunique': [df[col].nunique() for col in df.columns],
        'Cardinalidad_%': [(df[col].nunique() / total_filas) * 100 for col in df.columns],
        'Tipo': [df[col].dtype for col in df.columns]
    })

    reporte['Cardinalidad_%'] = reporte['Cardinalidad_%'].round(2)
    # Añadimos una recomendación rápida basada en el %
    def sugerir_tipo(row):
        if row['Cardinalidad_%'] == 100:
            return "Posible ID / Clave Primaria"
        elif row['Cardinalidad_%'] < 5:
            return "Categórica (Baja cardinalidad)"
        else:
            return "Numérica / Categórica compleja"
            
    reporte['Sugerencia'] = reporte.apply(sugerir_tipo, axis=1)
    
    return reporte.sort_values(by='Cardinalidad_%', ascending=False)



### FUNCIÓN 2: Obtiene las fechas máximas y mínimas de cada campo de cada tabla #####

def resumen_fechas(tablas: dict):
    """
    Recorre un diccionario de DataFrames y devuelve un DataFrame resumen
    con las columnas de tipo datetime y sus valores mínimo y máximo.
    
    Parámetros:
        tablas (dict): {"nombre_tabla": dataframe}
    
    Retorna:
        DataFrame con columnas:
        - tabla
        - campo
        - fecha_min
        - fecha_max
    """
    
    registros = []

    for nombre_tabla, df in tablas.items():
        # Recorremos columnas del DataFrame
        for columna in df.columns:
            # Detectamos columnas datetime
            if pd.api.types.is_datetime64_any_dtype(df[columna]):
                registros.append({
                    "tabla": nombre_tabla,
                    "campo": columna,
                    "fecha_min": df[columna].min(),
                    "fecha_max": df[columna].max()
                })

    return pd.DataFrame(registros)
```
